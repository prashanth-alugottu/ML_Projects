```python
# loading the dataset
news_dataset = pd.read_csv('news.csv')
print(news_dataset.shape)
```

```python
news_dataset = news_dataset.fillna('')
news_dataset.isnull().sum()
```

```python
X = news_dataset.drop(columns='real', axis=1)
X.head()
```

```python
Y = news_dataset['real']
Y.head()
```

 Steming proccesser

is the process of reducing the a word in its root words.
Ex: acting, actor, actress ==> act


```python
import re
from nltk.stem.porter import PorterStemmer
```

```python
port_stem = PorterStemmer()
```

```python
def stemming(sentense):
    stemming_text = re.sub('[^a-zA-Z]',' ',sentense)
    stemming_text = stemming_text.lower()
    stemming_text = stemming_text.split()
    stemming_text = [port_stem.stem(word) for word in stemming_text if not word in stopwords.words('english')]
    stemming_text = ' '.join(stemming_text)
    return stemming_text
```

```python
news_dataset['title'] = news_dataset['title'].apply(stemming)
news_dataset.head()
```

```python
# sepaarating the data and label
X  = news_dataset['title'].values
print(X)
print(X.shape)
```

```python
Y = news_dataset['real'].values
print(Y)
print(Y.shape)
```

Converting text data to n numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer() # Tfidf = Term Frequency Inverse Document Frequency
vectorizer.fit(X)               # Tf = how many times a word appears in a document (seems NLP conncept)
X = vectorizer.transform(X)     
print(X)
```

Spliiting data set to training ans test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size= 0.3, random_state = 42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the model : Logistic Regressin Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
# Accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on the training data : ", training_data_accuracy)
```

```python
# Accuracy on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("Accuracy on the test data : ", test_data_accuracy)
```
